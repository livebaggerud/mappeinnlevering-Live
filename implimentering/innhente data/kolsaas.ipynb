```python
import matplotlib.pyplot as plt

fil = open("kolsaas.csv")
data = fil.read()
```

```python
linjer = data.split("\n")
tur = []
for linje in linjer:
    splittet_linje = linje.split(",")
    tur.append(splittet_linje)

tur[0][1]
```

```python
x = []
y = []

lengde = []
bredde = []

for i in range(1, len(tur)):
    lengde.append(float(tur[i][2]))
    bredde.append(float(tur[i][1]))

for i in range(1,len(tur)):
    x.append(float(tur[i][0]))
    y.append(float(tur[i][3]))

plt.scatter(x,y)
plt.plot(lengde, bredde)
plt.show()
```
