```python
import matplotlib.pyplot as plt

#Oppgave 3

def f(x):
    return 2*x+3

x = []
y = []

for i in range(0,101):
    x.append(i)
    y.append(f(i))

plt.axhline(0, color="black", zorder=0)
plt.axvline(0, color = "black", zorder=0)

plt.title("$2*x+3$")
plt.xlabel("x")
plt.ylabel("y")

plt.plot(x,y)
plt.show()
```
