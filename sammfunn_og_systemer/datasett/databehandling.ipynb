### Datasett i Python

> En stor samling av data kalles ofte for et datasett

```python
import matplotlib.pyplot as plt

fil = open("./datasett/internet-users.csv")
data = fil.read()
linjer = data.split("\n")

brukere = []
år = []

for linje in linjer[:-1]:
    splittet_linje = linje.split(",")
    if splittet_linje[1] == "Norway":
        brukere.append(float(splittet_linje[6]))
        år.append(int(splittet_linje[3]))

plt.plot(år, brukere)
```

```python
### OPPGAVE 3


import matplotlib.pyplot as plt

fil = open("List of world cities by population density.csv")
data = fil.read()
linjer = data.split("\n")

land = []
for linje in linjer[:-1]:
    splittet_linje = linje.split(",")
    land.append(splittet_linje[-1])


høyeste = ""
høyeste_antall = -1
for ln in land:
    antall = land.count(ln)
    if antall > høyeste_antall:
        høyeste = ln
        høyeste_antall = antall

print(høyeste)


```

```python
#OPPGAVE 4

fil = open("AirBnb-Europe.csv")
data = fil.read()
linjer = data.split("\n")

priser = []
byer = []

for linje in linjer[1:-1]:
    splittet_linje = linje.split(",")
    priser.append(float(splittet_linje[1]))
    byer.append(splittet_linje[0])
    
laveste = 100000000
for pris in priser:
    if pris < laveste:
        laveste = pris

print(laveste)
```
